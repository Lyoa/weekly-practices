## Problem No. 1 (10pts.)

Create a function that will accept two numbers. These numbers will represent a position on a tic-tac-toe board. It should be 0 to 8, where 0 is the top-left spot, and 8 is the bottom-right spot.

These parameters are two marks on the tic-tac-toe board. The function should return the representation of the board for the spot that can block these two spots, which is represented as 'O', from winning the game using 'X' (literal string). If the two given numbers cannot win the game in one more move, the function should return a message about this instead of showing the board.

CONSTRAINT:<br>
&emsp;&emsp;• Do not just literally print the board pattern<br>
&emsp;&emsp;• Make use of the loop statement to produce the board and to derive patterns from the parameters of the function<br>

INPUT (1):<br>
find_blocking_spot(0,1)<br>

OUTPUT (1):<br>
&ensp;O O X<br>
&nbsp;&ensp;\*&ensp;\*&ensp;\*<br>
&nbsp;&ensp;\*&ensp;\*&ensp;\*<br>

INPUT (2):<br>
find_blocking_spot(2,8)<br>

OUTPUT (2):<br>
&nbsp;&ensp;\*&ensp;\*&ensp;O<br>
&nbsp;&ensp;\*&ensp;\*&ensp;X<br>
&nbsp;&ensp;\*&ensp;\*&ensp;O<br>

INPUT (3):<br>
find_blocking_spot(4,6)<br>

OUTPUT (3):<br>
&nbsp;&ensp;\*&ensp;\*&ensp;X<br>
&nbsp;&ensp;\*&ensp;O&ensp;\*<br>
&nbsp;&ensp;O&ensp;\*&ensp;\*<br>

INPUT (4):<br>
find_blocking_spot(0,5)<br>

OUTPUT (4):<br>
These two spots can't win the game in one more move

INPUT (5):<br>
find_blocking_spot(1,3)<br>

OUTPUT (5):<br>
These two spots can't win the game in one more move

```python
#Problem 1
def find_blocking_spot(x, y):
    winning_combi = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]
    combi = [combi for combi in winning_combi if x in combi and y in combi]
    combi_board = ['*', '*', '*', '*', '*','*', '*', '*', '*',]

    if combi == []:
        print ("These two spots can\'t win the game in one more move")
    else:
        for z in combi: # 1st
            for z_bel in z: #2nd
                if x == z_bel:    #condi        
                    combi_board [int(x)] = 'O'
                elif y == z_bel:
                    combi_board [int(y)] = 'O'
                else:
                    combi_board [int(z_bel)] = 'X'
            #for every i and x in enumerate(combi_board):
            for i, x in enumerate(combi_board):
                print(x, end = '')
                if i in (2, 5, 8):
                    print()

find_blocking_spot(1, 3)
```
