## Problem No. 1

You have been hired by a car rental company to create a program to manage their fleet of cars. Each car has a make, model, year, and rental price per day. There are two types of cars: economy and luxury.

Economy cars have a rental price of $50 per day, while luxury cars have a rental price of $100 per day.

Create a Car class with attributes for make, model, year, and rental price per day. Then, create two subclasses of Car: EconomyCar and LuxuryCar.

The EconomyCar subclass should inherit from the Car class and set the rental price per day to $50. The LuxuryCar subclass should inherit from the Car class and set the rental price per day to $100.

Your program should allow the user to enter information for a car (make, model, year) and specify whether it is an economy or luxury car. The program should then calculate and display the total rental cost for a given number of days.

[Example]

Enter make: Toyota<br>
Enter model: Corolla<br>
Enter year: 2022<br>
Is this car an economy car (y/n)? y<br>
Enter number of rental days: 5<br>

Total rental cost: $250<br>

[When the string representation of the Car class object is printed, the output should be:]

print(car) # 'car' variable should be an instance of either of the two subClass.

make: Toyota<br>
model: Corolla<br>
year: 2022<br>

[Constraints]

- All input values are strings or integers.
- The rental price per day for economy cars is $50 and for luxury cars is $100.
- The number of rental days entered by the user is a positive integer.


```python
# Problem 1
#parent
class Car:
    def __init__(self, make, model, year, rental_price):
        self.make = make
        self.model = model
        self.year = year
        self.rental_price = rental_price
    
    def __str__(self):
        return f"make: {self.make}\nmodel: {self.model}\nyear: {self.year}"

#subclass1
class EconomyCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year, 50)
    
 #subclass2   
class LuxuryCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year, 100)

def car_info():
    make = input("Enter make: ")
    model = input("Enter model: ")
    year = int(input("Enter year: "))
    while True:
        is_economy = input("Is this car an economy car (y/n)? ").lower()
        if is_economy == 'y':
            car = EconomyCar(make, model, year)
            break
        elif is_economy == 'n':
            car = LuxuryCar(make, model, year)
            break
        else:
            print("Invalid input. Please enter 'y' or 'n'.")
    return car

def rental_cost(car, days):
    return car.rental_price * days 

if __name__ == '__main__':
    car = car_info()
    days = int(input("Enter number of rental days: "))
    total_cost = rental_cost(car, days)
    print(f"\nTotal rental cost: ${total_cost}")
    
    print(car)
```

## Problem No. 2
Create a class named Animal with the following attributes and methods:

Attributes:<br>

name (string): the name of the animal<br>
age (integer): the age of the animal<br>
weight (float): the weight of the animal<br>

Methods:<br>

__init__(self, name, age, weight): constructor method that initializes the name, age, and weight attributes<br>
eat(self, food): method that takes in a food parameter and prints a message saying that the animal is eating that food<br>
sleep(self): method that prints a message saying that the animal is sleeping<br>
Create a subclass of Animal named Dog with the following additional attributes and methods:<br>

Attributes:<br>

breed (string): the breed of the dog<br>
color (string): the color of the dog<br>

Methods:<br>

__init__(self, name, age, weight, breed, color): constructor method that initializes the name, age, weight, breed, and color attributes<br>
bark(self): method that prints a message saying that the dog is barking<br>
Create an instance of the Dog class and demonstrate the use of its attributes and methods.<br>

[Input]
my_dog = Dog("Max", 3, 25.5, "Golden Retriever", "Golden")<br>
print(f"{my_dog.name} is a {my_dog.breed} dog with {my_dog.color} color, {my_dog.age} years old, and {my_dog.weight} kg weight.")<br>
my_dog.bark()<br>
my_dog.eat("dog food")<br>
my_dog.sleep()<br>

[Output]
Max is a Golden Retriever dog with Golden color, 3 years old, and 25.5 kg weight.<br>
Max is barking.<br>
Max is eating dog food.<br>
Max is sleeping.<br>

```python
#Problem 2
class Animal:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight
        
class Dog(Animal):
    def __init__(self, name, age, weight, breed, color):
        super().__init__(name, age, weight)
        self.breed = breed
        self.color = color
        
    def bark(self):
        print (self.name + ' is barking')

    def sleep(self):
        print (self.name + ' is sleeping')

    def eat(self, food):
        print (self.name + f' is eating {food}')
    
my_dog = Dog("Max", 3, 25.5, "Golden Retriever", "Golden")
print(f"{my_dog.name} is a {my_dog.breed} dog with {my_dog.color} color, {my_dog.age} years old, and {my_dog.weight} kg weight.")
my_dog.bark()
my_dog.eat("dog food")
my_dog.sleep()
```
